```python
import pandas as pd
import numpy as np
```

```python
filmes = pd.read_csv('movies_metadata.csv', low_memory=False)
filmes.head(5)
```

```python
avaliacoes = pd.read_csv('ratings.csv')
avaliacoes.head()
```

```python
 
filmes = filmes [['id','original_title','original_language','vote_count']]
 
filmes.rename(columns = {'id':'ID_FILME','original_title':'TITULO','original_language':'LINGUAGEM','vote_count':'QT_AVALIACOES'}, inplace = True)

filmes.head()
```

```python
avaliacoes = avaliacoes[['userId','movieId','rating']]
avaliacoes.rename(columns = {'userId':'ID_USUARIO', 'MOVIEiD':'FILME_ID', 'rating':'AVALIACAO'}, inplace = True)
avaliacoes.head()
```

```python

avaliacoes.rename(columns = {'movieId':'FILME_ID'}, inplace = True)
avaliacoes.head()
```

```python

avaliacoes.rename(columns = {'FILME_ID':'ID_FILME'}, inplace = True)
avaliacoes.head()
```

```python
filmes.isna()
```

```python
filmes.isna().sum()
```

```python
filmes.dropna(inplace = True)
```

```python
filmes.isna().sum()
```

```python
avaliacoes['ID_USUARIO'].value_counts()
```

```python
qt_avaliacoes = avaliacoes['ID_USUARIO'].value_counts() > 500
```

```python
y = qt_avaliacoes[qt_avaliacoes].index
y.shape
```

```python
avaliacoes.shape
```

```python
avaliacoes = avaliacoes[avaliacoes['ID_USUARIO'].isin(y)]
avaliacoes.shape
```

```python
avaliacoes.head()
```

```python
filmes.head()
```

```python
filmes = filmes[filmes['QT_AVALIACOES'] > 500]
```

```python
filmes.head()
```

```python
filmes_lang = filmes['LINGUAGEM'].value_counts()
filmes_lang.head(10)
```

```python
filmes.info()
```

```python
avaliacoes.info()
```

```python
filmes['ID_FILME'] = filmes['ID_FILME'].astype(int)
```

```python
filmes.shape
```

```python
filmes.info()
```

```python
filmes_e_avaliacoes = avaliacoes.merge(filmes, on = 'ID_FILME')
filmes_e_avaliacoes.head()
```

```python
filmes_e_avaliacoes.isna().sum()
```

```python
filmes_e_avaliacoes.head(20)
```

```python
filmes_e_avaliacoes.drop_duplicates(['ID_USUARIO','ID_FILME'], inplace = True)
```

```python
filmes_e_avaliacoes.shape
```

```python
del filmes_e_avaliacoes['ID_FILME']
filmes_e_avaliacoes.head()
```

```python
filmes_p = filmes_e_avaliacoes.pivot_table(columns = 'ID_USUARIO', index = 'TITULO', values = 'AVALIACAO')
```

```python
filmes_e_avaliacoes.head(20)
```

```python
filmes_p.head(20)
```

```python
filmes_p.fillna(0, inplace = True)
filmes_p.head(20)
```

```python
from scipy.sparse import csr_matrix
```

```python
films_parsed = csr_matrix(filmes_p)
```

```python
type(films_parsed)
```

```python
from sklearn.neighbors import NearestNeighbors
modelo = NearestNeighbors(algorithm = 'brute')
modelo.fit(films_parsed)
```

```python
 # testando com o filme Aliens vs Predator: Requiem
distances, sugestions = modelo.kneighbors(filmes_p.filter(items = ['Aliens vs Predator: Requiem'], axis=0).values.reshape(1, -1))

for i in range(len(sugestions)):
    print(filmes_p.index[sugestions[i]])  
```

```python
 # testando com o filme Avengers: Age of Ultron	
distances, sugestions = modelo.kneighbors(filmes_p.filter(items = ['Avengers: Age of Ultron'], axis=0).values.reshape(1, -1))

for i in range(len(sugestions)):
    print(filmes_p.index[sugestions[i]])  
```
